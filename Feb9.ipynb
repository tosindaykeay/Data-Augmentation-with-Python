```python
!pip install transformers gradio torch

from transformers import pipeline
import gradio as gr

# Load primary moderation model
moderation_pipeline = pipeline("text-classification", model="facebook/roberta-hate-speech-dynabench-r4-target")

# Load a secondary sentiment analysis model for better accuracy
sentiment_pipeline = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")

# Function to moderate text more accurately
def moderate_text(user_text):
    # Get moderation result
    mod_result = moderation_pipeline(user_text)[0]
    mod_label = mod_result['label']
    mod_score = mod_result['score']

    # Get sentiment result for more context
    sent_result = sentiment_pipeline(user_text)[0]
    sent_label = sent_result['label']
    sent_score = sent_result['score']

    # Classification Logic
    if mod_label == "hate-speech" and mod_score > 0.8:
        return f"🚨 **Hate Speech Detected!**\nConfidence: {mod_score:.2f}\n⚠️ Please reconsider your words!"
    elif mod_label == "hate-speech" and 0.5 < mod_score <= 0.8:
        return f"⚠️ **Potentially Offensive Content**\nConfidence: {mod_score:.2f}\n🤔 This message may be problematic."
    elif sent_label == "NEGATIVE" and sent_score > 0.75:
        return f"🛑 **Negative Sentiment Detected**\nConfidence: {sent_score:.2f}\nConsider rewording for a friendlier tone."
    else:
        return f"✅ **Safe Content**\nConfidence: {mod_score:.2f}\n👍 No issues detected!"

# Gradio UI
iface = gr.Interface(
    fn=moderate_text,
    inputs=gr.Textbox(placeholder="Type your message...", label="Text Input"),
    outputs=gr.Textbox(label="Moderation Result"),
    title="🔥 Advanced Text Moderator",
    description="🚀 This AI detects hate speech, offensive language, and negative sentiment. Type something to check its safety!",
    theme="compact"
)

# Launch the app
iface.launch(share=True)
```
